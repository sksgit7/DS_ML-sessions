## Base-N encoding

This is comparatively better to encode features with large number of nominal type unique values.<br>
<i>Binary encoding</i> the integer values assigned to unique elements are converted into combination of 0 and 1 as the base is 2. Here the digits representing the values depend on the <i>base</i> of encoding.<br><br>

If base=4, then the digits can be 0,1,2,3. So, if the 10th unique value is numbered 10, then it can be encoded as (0 2 2), which requires only 3 features. But in <i>Binary encoding</i> it would have been encoded as (1 0 1 0).<br><br>

So, it is a good method to encode with less number of extra features when the unique categorical values are more in a column

```
import category_encoders as ce
df=pd.DataFrame({'letters':['a','e','h','b','c','d','d','h','i','f','g','j','g','g']})
df
```

```
encoder= ce.BaseNEncoder(cols=['letters'],base=4)
df1= encoder.fit_transform(df)
df1
```

Here 'j' is the 10th unique letter, encoded as (0 2 2)
